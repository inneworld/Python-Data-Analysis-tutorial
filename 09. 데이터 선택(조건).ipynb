# 9. 데이터 선택(조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df['키'] >= 185 # 185 이상이면 True 아니면 False
```

```python
filt = (df['키'] >= 185)
df[filt]
```

```python
df[~filt] # filter를 역으로 적용
```

```python
df[df['키'] >= 185] # 변수로 지정하지 않고도 가능
```

```python
df.loc[df['키'] >=185, '수학'] # 이런식으로 원하는 col을 가져올 수 있음
```

```python
df.loc[df['키'] >=185, ['수학','과학']]
```

## 다양한 조건

### & 그리고

```python
df.loc[(df['키'] >= 185) &  (df['학교'] == '북산고')] # 키가 185이상인 북산고 학생
```

### | 또는

```python
df.loc[(df['키'] >= 185) |  (df['학교'] == '북산고')]# 키가 185이상 또는 북산고 학생
```

### str 함수

```python
filt = df['이름'].str.startswith('송') # 송으로 시작하는 사람
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 이름에 '태' 가 들어간 사람
df[filt]
```

```python
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs) # sw특기가 python 또는 java인 사람
df[filt]
```

```python
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs) # (대소문자 무시)
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java', na = False) # NaN 데이터는 False로 지정
df[filt] # Java라는 글자를 포함하는 데이터 출력
```
