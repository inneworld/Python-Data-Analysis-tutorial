# 3. 범례 (legend)

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 맑은 고딕 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시 마이너스 깨짐해결 
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y,label='무슨 데이터')
plt.legend()
```

```python
plt.plot(x,y,label='무슨 데이터') # 위치 옮기기
plt.legend(loc='upper right') # 오른쪽 위
```

```python
plt.plot(x,y,label='무슨 데이터') 
plt.legend(loc=(0.7,0.8)) # 이런식으로 숫자로 지정도 가능
```
