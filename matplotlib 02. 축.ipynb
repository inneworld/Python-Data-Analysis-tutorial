# 2. 축

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 맑은 고딕 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시 마이너스 깨짐해결 
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프')
```

### 축 설정

```python
plt.plot(x,y)
plt.xlabel('X 축') # 축에 이름 짓기
plt.ylabel('Y 축')
```

```python
plt.plot(x,y)
plt.xlabel('X축',color='red',loc='right') # color 는 색을 설정 , loc 는 정렬 기준
plt.ylabel('Y축',color='green',loc='top')
```

```python
plt.plot(x,y)  # 축에 어떤 값들을 표시할것인지
plt.xticks([1,2,3]) 
plt.yticks([3,6,9,12])
```
