# 9. 막대그래프 (심화)

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 맑은 고딕 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시 마이너스 깨짐해결
```

```python
labels = ['강백호','서태웅','정대만']
values=[190,187,184]
plt.barh(labels, values) # 가로 막대 그래프
plt.xlim(175,195)
```

```python
bar = plt.bar(labels,values) 
bar[0].set_hatch('/') # 이런식으로 표현 가능
bar[1].set_hatch('*')
bar[2].set_hatch('+')
```

```python
bar = plt.bar(labels,values)
plt.ylim(175,195)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() + 0.5, values[idx])
```
