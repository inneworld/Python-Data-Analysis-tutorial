# 13. 원그래프 (기본)

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 맑은 고딕 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시 마이너스 깨짐해결
```

```python
values = [30 , 25 , 20 , 13 , 10 , 21]
labels = ['python','java','javascript','c#','c/c++','etc']

plt.pie(values,labels = labels)
plt.show()
```

## 퍼센트 표시

```python
values = [30 , 25 , 20 , 13 , 10 , 21]
labels = ['python','java','javascript','c#','c/c++','etc']

plt.pie(values,labels = labels,autopct='%.1f')
plt.show()
```

```python
values = [30 , 25 , 20 , 13 , 10 , 21]
labels = ['python','java','javascript','c#','c/c++','etc']

plt.pie(values,labels = labels,autopct='%.1f%%')
plt.show()
```

## 그래프 시작 각도

```python
values = [30 , 25 , 20 , 13 , 10 , 21]
labels = ['python','java','javascript','c#','c/c++','etc']

plt.pie(values,labels = labels,autopct='%.1f%%',startangle = 90) #  그래프 시작 각도
plt.show()
```

## 그래프 방향

```python
values = [30 , 25 , 20 , 13 , 10 , 21]
labels = ['python','java','javascript','c#','c/c++','etc']

plt.pie(values,labels = labels,autopct='%.1f%%',counterclock = False) # 방향 (반 시계 방향)
plt.show()
```

## 그래프 띄우기

```python
values = [30 , 25 , 20 , 13 , 10 , 21]
labels = ['python','java','javascript','c#','c/c++','etc']
explode = [0.05] * 6
plt.pie(values , labels = labels , explode = explode)
plt.show()
```

## 범례

```python
plt.pie(values , labels = labels , explode = explode)
plt.legend(loc =(1.2 , 0.3)) # 위치를 좌표로 이동 
plt.show()
```

```python
plt.pie(values , labels = labels , explode = explode)
plt.legend(loc =(1.2 , 0.3) , title = '언어별 선호도') 
plt.show()
```
