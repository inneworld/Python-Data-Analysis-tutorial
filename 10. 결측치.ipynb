## 데이터 제외하기 dropna

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
df.dropna() # NaN 을 포함하는 데이터를 삭제
```

- axis :index or columns  
열을 제거 할건지 , 행을 제거 할건지 선택
- how : any or all   
한개만 있어도 삭제 할건지 전체가 nan일때 삭제 할건지

```python
df.dropna(axis='index', how='any') # nan이 포함된 index를 삭제 any이기 때문에 하나만 nan이어도 삭제가 된다.
```
