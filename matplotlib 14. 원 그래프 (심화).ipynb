# 14. 원 그래프 (심화)

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 맑은 고딕 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시 마이너스 깨짐해결
```

```python
values = [30 , 25 , 20 , 13 , 10 , 21]
labels = ['python','java','javascript','c#','c/c++','etc']

plt.pie(values,labels = labels)
plt.show()
```

### 색 바꾸기

```python
colors = ['b','g','r','c','m','y']
plt.pie(values,labels = labels, colors = colors)
plt.show
```

### 중간 비게 하기

```python
wedgprops={'width':0.6}
plt.pie(values , labels = labels , wedgeprops=wedgprops)
plt.show()
```

### 특정조건만 출력하기

```python
def custom_autopct(pct):
    return ('%.1f%%' % pct) if pct >=10 else '' # 10 퍼센트 이하는 퍼센트 출력 X

plt.pie(values , labels = labels , wedgeprops= wedgprops , autopct = custom_autopct)
plt.show()
```
