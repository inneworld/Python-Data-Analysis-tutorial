# 4. 스타일

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 맑은 고딕 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시 마이너스 깨짐해결 
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth = 10) # 선 두께
```

## marker

```python
plt.plot(x,y,marker='o') # 마커 종류 고를 수 있음
```

```python
plt.plot(x,y,marker='v', markersize = 10) # 마커 사이즈 설정
```

```python
plt.plot(x,y,marker='X', markersize=10 , markeredgecolor='red') # 마커 테두리 색상
```

```python
plt.plot(x,y,marker='o', markersize=20,markerfacecolor='yellow') # 마커 색상
```

### 선 스타일

```python
plt.plot(x,y,linestyle = ':') # 점선
```

```python
plt.plot(x,y,linestyle = '--') 
```

```python
plt.plot(x,y,linestyle = '-.') 
```

### 색상

```python
plt.plot(x,y,color = 'pink') # 선 색상 설정
```

```python
plt.plot(x,y,color='#ff0000') # rgb 로도 설정 가능
```

### 포멧

```python
plt.plot(x,y,'ro--') # color , marker, linestyle 순서로 한번에 지정도 가능
```

### 축약어

```python
plt.plot(x,y,marker = 'o', mfc='red', ms = 10 , mec = 'yellow' , ls = ':') 
# mfs = macker face color ,ms = marker size ,mec = marker edge color ,ls = line style
```

### 투명도

```python
plt.plot(x,y,marker='o', mfc ='red', ms=10, alpha = 1) # alpha 가 투명도 설정
```

### 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5), dpi=150) # dpi 로 해상도 설정 가능
plt.plot(x,y)
```

### 배경색

```python
plt.figure(facecolor = 'yellow')
plt.plot(x,y)
```
